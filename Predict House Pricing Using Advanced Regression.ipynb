```python
from sklearn.linear_model import LinearRegression      # import packages.
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, random_state = 42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.metrics import r2_score, mean_squared_error,mean_absolute_error
from sklearn.linear_model import LinearRegression, ElasticNet, Ridge, Lasso
from sklearn.tree import ExtraTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
lr = LinearRegression()
```

```python
model = lr.fit(x,y)
```

```python
tahmin = model.predict(x_test)
```

```python
(mean_squared_error(tahmin, y_test))**.5    
```

```python
r2_score(tahmin, y_test)  
```

```python
R = Ridge()
R.fit(x_train,y_train)
y_pred_ridge = R.predict(x_test) 
```

```python
(mean_squared_error(y_test,y_pred_ridge))**0.5
```

```python
r2_score(y_test,y_pred_ridge)
```

```python
ETR = ExtraTreeRegressor()
ETR.fit(x_train,y_train)
y_pred_etr = ETR.predict(x_test) 
```

```python
(mean_squared_error(y_test,y_pred_etr))**0.5  
```

```python
r2_score(y_test,y_pred_etr)
```
